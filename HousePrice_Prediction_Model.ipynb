```python
# Check which cols are having a lot of missing data after that we can remove that columns

#checking the columns with missing data
Miss_Percent=100*(train_data.isnull().sum()/len(train_data))

#Sorting the data columns by their percentage in descending order
Miss_Percent=Miss_Percent[Miss_Percent>0].sort_values(ascending=False).round(1)

#Creating a dataframe to show percentage of missing data and its respective data column in table
DataFrame=pd.DataFrame(Miss_Percent)
miss_percent_table=DataFrame.rename(columns={0:'Percentage of Missing Values'})
MissPercent=miss_percent_table

#Displaying Missing Value table
MissPercent
```

### Preparing for Modelling.

```python
#Dropping Columns because bescause they were dropped in Training data
test_id=test_data['Id']
test_data=test_data.drop(['Id','Alley','PoolQC','MiscFeature','Fence'],axis=1)
```

```python
#Converting String into Number
# As Machine can only deal with 0-1 we need to convert string into numerals
new_train_data=train_data.apply(lambda x:pd.factorize(x)[0])
test_data=test_data.apply(lambda x:pd.factorize(x)[0])
```

```python
from sklearn.model_selection import train_test_split
X = new_train_data.drop('SalePrice' , axis = 1)
Y = new_train_data['SalePrice']
X_train , X_valid , Y_train , Y_valid = train_test_split(X , Y , test_size=0.2 , random_state=40 , shuffle=True)
```

```python
#Allocating training set and testing set
# Now devide dataset into training and validation datasets
X_test=test_data
X_train.shape, Y_train.shape, X_valid.shape, Y_valid.shape, X_test.shape
```

## Code for applying Linear Regression

```python
# Use linear regression & fit training dataset
linearModel=LinearRegression()
linearModel.fit(X_train, Y_train)
```

```python
#code for making predictions using Linear Model on validation dataset
from sklearn import metrics
print('Training score : {}'.format(linearModel.score(X_train,Y_train)))
Prediction=linearModel.predict(X_valid)
r2_score = metrics.r2_score(Y_valid , Prediction)
print('r2 score: {}'.format(r2_score))
print('RMSE:',np.sqrt(metrics.mean_squared_error(Y_valid , Prediction)))
sns.distplot(Y_valid - Prediction)
```

```python
Linear_predict=linearModel.predict(X_test)
```

```python
#Creating Dataframe to store the Ids with Prediction
linear_output=pd.DataFrame({'Id':test_id,'SalePrice':Linear_predict})
print(linear_output)
```
